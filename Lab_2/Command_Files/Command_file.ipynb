```python
# Imports
from matplotlib import pyplot as plt
import pandas as pd

# Read data from .csv
data = pd.read_csv('C:/Users/Dell/Documents/Studia_sem7/Analiza-i-Bazy-Danych/Lab_2/Original_Data/drinks.csv')

# Show original data
data
```

```python
# Deleting empty rows
data = data[(data[['total_litres_of_pure_alcohol']] != 0).all(axis=1)]

# Save to .csv file
data.to_csv('C:/Users/Dell/Documents/Studia_sem7/Analiza-i-Bazy-Danych/Lab_2/Analysis_Data/valid_data.csv')

# Show data without empty rows
data
```

```python
beer_servings = data[['country','beer_servings']]
beer_servings = beer_servings.sort_values('beer_servings',ascending=False)
beer_servings = beer_servings.head(10)

# Save to .csv file
beer_servings.to_csv('C:/Users/Dell/Documents/Studia_sem7/Analiza-i-Bazy-Danych/Lab_2/Analysis_Data/top10_beer_servings.csv')

beer_servings
```

```python
beer_servings.plot.bar(x='country', y='beer_servings')
plt.title("Top 10 countries by beer servings consumed per person")
plt.ylabel("servings")
#plt.xticks(rotation=90)

#save to file 
plt.savefig('C:/Users/Dell/Documents/Studia_sem7/Analiza-i-Bazy-Danych/Lab_2/Analysis_Data/top10_beer_servings')
```
