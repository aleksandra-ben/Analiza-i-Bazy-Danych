```python
# Imports
from matplotlib import pyplot as plt
import pandas as pd

# Read data from .csv
data = pd.read_csv('C:/Users/Dell/Analiza-i-Bazy-Danych/Ćwiczenie_1-TIER_protocol_i_tidy_data/Original Data/drinks.csv')

# Show original data
data
```

```python
# Deleting empty rows
data = data[(data[['total_litres_of_pure_alcohol']] != 0).all(axis=1)]

# Show data without empty rows
data
```

```python
beer_servings = data[['country','beer_servings']]
beer_servings = beer_servings.sort_values('beer_servings',ascending=False)
beer_servings = beer_servings.head(10)
beer_servings
```

```python
beer_servings.plot.bar(x='country', y='beer_servings', rot=0)
plt.title("Top 10 countries by beer servings consumed per person")
plt.ylabel("servings")
```

```python
spirit_servings = data[['country','spirit_servings']]
spirit_servings = spirit_servings.sort_values('spirit_servings',ascending=False)
spirit_servings = spirit_servings.head(10)
spirit_servings
```

```python
spirit_servings.plot.bar(x='country', y='spirit_servings', rot=0)
plt.title("Top 10 countries by spirit servings consumed per person")
plt.ylabel("servings")
```

```python
wine_servings = data[['country','wine_servings']]
wine_servings = wine_servings.sort_values('wine_servings',ascending=False)
wine_servings = wine_servings.head(10)
wine_servings
```

```python
wine_servings.plot.bar(x='country', y='wine_servings', rot=0)
plt.title("Top 10 countries by wine servings consumed per person")
plt.ylabel("servings")
```

```python
total_litres_of_pure_alcohol = data
total_litres_of_pure_alcohol = total_litres_of_pure_alcohol.sort_values('total_litres_of_pure_alcohol',ascending=False)
total_litres_of_pure_alcohol = total_litres_of_pure_alcohol.head(10)
total_litres_of_pure_alcohol
```

```python
total_litres_of_pure_alcohol.plot.bar(x='country', y={'beer_servings','spirit_servings','wine_servings'})
plt.title("Top 10 countries by total litres of pure alcohol consumed per person")
plt.ylabel("servings")
```
